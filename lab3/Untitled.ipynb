```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['tip'], bins=10)
#---------------------------------
#kde - bool for distribution curve
#bins - number of columns
#---------------------------------
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')

#kind - scatter, reg, resid, kde, hex
```

```python
sns.pairplot(tips, hue='sex')
```

```python
import matplotlib.pyplot as plt
import numpy as np
bplot = sns.barplot(x='sex',y='total_bill',data=tips,hue='tip', estimator = np.median)
bplot.legend_.remove()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette = 'twilight')
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette = 'Set2')
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
grid = sns.FacetGrid(tips, col='time', row='smoker', hue='sex', palette = sns.color_palette(["#34495e","#e74c3c"]))
grid = grid.map(plt.hist, 'tip')
grid.add_legend()
```
