#### 1. Основи побудови графіків

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

#### 2. Побудова графіків

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#### 2.2 Об'єктно-орієнтований підхід

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Title')

axes2.plot(x,y,'b')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4)
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x^2")
ax.plot(x,x**3,label="x^3")
ax.legend()
```

```python
fig = plt.figure(figsize = (8,4), dpi = 72)
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,color = 'b', marker = 'o')
ax.plot(x,x**3,color = 'g', lw = 2, ls = '--')
```

```python
fig.savefig('figure1.png')
```
